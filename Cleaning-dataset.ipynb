```python
import pandas as pd
import numpy as np
```

```python
FILE_NAME = "dataset.handphone.csv"

FEATURES_TO_COMPARE = [
    'Harga', 
    'Ram', 
    'Memori_internal', 
    'Resolusi_kamera', 
    'Kapasitas_baterai', 
    'Brand', 
    'Os'
]

NUMERIC_FEATURES = ['Harga', 'Ram', 'Memori_internal', 'Resolusi_kamera', 'Kapasitas_baterai']
CATEGORICAL_FEATURES = ['Brand', 'Os']
```

```python
df = pd.read_csv(FILE_NAME, delimiter=";")
print("Data loaded successfully!")
```

```python
df['Resolusi_kamera'] = df['Resolusi_kamera'].astype(str).str.extract(r'(\d+)').astype(float)

print("\n--- 'Resolusi_kamera' (After Cleaning) ---")
print(df['Resolusi_kamera'].head())
```

```python
for col in NUMERIC_FEATURES:
    # Kita juga perlu menangani jika 'Harga' memiliki 'Rp' atau '.'
    if col == 'Harga':
        # Ubah ke string, hapus 'Rp' dan '.', lalu ubah ke numerik
        # (Jika format harga Anda bersih, Anda bisa lewati ini)
        df[col] = df[col].astype(str).str.replace('Rp', '').str.replace('.', '', regex=False)
        pass # Diasumsikan data harga sudah bersih seperti di file
        
    df[col] = pd.to_numeric(df[col], errors='coerce')

print("\n--- Data Info (After Numeric Conversion) ---")
df.info()
```
